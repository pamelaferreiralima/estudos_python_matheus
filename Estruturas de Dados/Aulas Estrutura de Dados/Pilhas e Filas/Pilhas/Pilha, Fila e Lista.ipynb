### Pilhas
1. Utilizado para empilhar e depois desempilhar
2. Quando vamos desempilhar as Pilhas, teremos um novo topo a cada retirada dos elementos. Fornecendo as últimas modificações realizadas.
3. Tipo abstrato de dados. O processador não 'entende' de forma direta, havendo uma conversão para tipos primitivos (mais simples) >>> Classe Stack (**Molde**); As classes auxiliam a mantermos os métodos que podem ser utilizados para seguir a regra das Pilhas (Cria objetos de classe)
3.1. Classe - Atributos - Operações
4. As Pilhas nada mais _**são do que listas que seguem uma regra (*LIFO*)**_
5. Operações/Métodos: Stack(), push(item), pop(), peek(), is_empt(), size()

Toda classe tem um método construtor, que é utilizado para construir um objeto daquela classe específica (*Operação*); Tem o mesmo nome da classe. Em Python não temos o mesmo nome da classe, temos um método chamado __init__(self)

```python
class Stack:
    def __init__(self): # método construtor
        self.itens = [] # criamos como referência para a classe uma lista vazia.
        
    def is_empty(self):
        return self.itens == []
    
    def pop(self):
        return self.itens.pop()
    
    def push(self, item):
        self.itens.append(item)
        
    def peek(self):
        return self.itens[len(self.itens) - 1]
    
    def size(self):
        return len(self.itens)
```

```python
# utilizando o exemplo das primeiras aulas, podemos fazer aquele algoritmo que converte para número binário.
def divide_2(dec_number):
    rem_stack = Stack()
    
    while dec_number > 0: # realizamos as operações para a divisão inteira
        rem = dec_number % 2
        rem_stack.push(rem)
        dec_number = dec_number // 2
        
    bin_string = "" # iniciamos a variável string vazia
    while not rem_stack.is_empty(): # enquanto o valor da Pilha não estiver vazia
        bin_string = bin_string + str(rem_stack.pop()) # desempilhamos e concatenamos os valores em bin_string
            
    return bin_string # retornamos o valor da string
```

```python
print(divide_2(100), divide_2(11500), divide_2(55888554))
```

```python
# converter de infixo para posfixo

def infix_to_postfix(infix_expr):
    prec = {} # preciso saber a precedência. Vincula os operadores com os valores (pesos)
    # dando peso as operações
    prec['*'] = 3
    prec['/'] = 3
    prec['+'] = 2
    prec['-'] = 2
    prec['('] = 1
    op_stack = Stack() #Criando uma variável do tipo pilha. Add os itens relacionados as operações da expressão
    # print(op_stack)
    postfix_list = [] # Lista que irá armazenar os itens do posfixo.
    token_list = infix_expr.split() # o splite irá 'quebrar' em unidades a expressão inteira
                                    # Podemos utilizar o splite porque a expressão inflix_expr é uma string
    # tratamos a expressão nesse bloco de código
    for token in token_list:
        # print(token)
        if token in "ABCDEFGHIJKLMNOPQRSTUVXYZ" or token in '0123456789':
            postfix_list.append(token)
            # print("primeiro if")
            # TRATAMOS OS PARENTESES. NA NOTAÇÕA POSFIXA NÃO TEMOS PARENTESES!
        elif token == "(": # ao encontrar um '(' add esse item a Pilha
            op_stack.push(token)
            # print("primeiro elif - ( ")
        elif token == ")": # precisamos agora encontrar o que fecha o parênteses ')'
            top_token = op_stack.pop()
            # print("segundo elif - ) ")
            while top_token != '(':
                postfix_list.append(top_token)
                top_token = op_stack.pop()
        else:
            # print("else") # TRATANDO OS OPERADORES
            while (not op_stack.is_empty()) and (prec[op_stack.peek()] >= prec[token]):
                postfix_list.append(op_stack.pop())
            op_stack.push(token)
    
    # Olhando o que temos na pilha após terminar toda a expressão
    while not op_stack.is_empty():
        postfix_list.append(op_stack.pop())
    return " ".join(postfix_list) # O .join (irá juntar todos em um formato de lista)
```

```python
print(infix_to_postfix(" ( A + B ) * C * ( D + E ) * ( F + G ) " )) # <__main__.Stack object at 0x0000022E26190F70>
```

```python
# Completed extended par_checker for: [,{,(,),},]
def par_checker(symbol_string):
    s = Stack()
    balanced = True
    index = 0
    while index < len(symbol_string) and balanced:
        symbol = symbol_string[index]
        if symbol in "([{":
            s.push(symbol)
        else:
            if s.is_empty():
                balanced = False
            else:
                top = s.pop()
                if not matches(top, symbol):
                    balanced = False
        index = index + 1
    if balanced and s.is_empty():
        return True
    else:
        return False

def matches(open, close):
    opens = "([{"
    closes = ")]}"
    # comparação feita pelo indice de cada string (o metodo index utiliza
    # como parametro o valor do símbolo para retornar o seu indice)
    return opens.index(open) == closes.index(close)
```

```python

print(par_checker('{{([][])}()}'))
print(par_checker('[{()]'))
```

```python
def postfix_eval(postfix_expr):
    operand_stack = Stack()
    token_list = postfix_expr.split()
    for token in token_list:
        if token in "0123456789":
            operand_stack.push(int(token))
        else:
            operand2 = operand_stack.pop()
            operand1 = operand_stack.pop()
            result = do_math(token, operand1, operand2)
            operand_stack.push(result)
    return operand_stack.pop()


def do_math(op, op1, op2):
    if op == "*":
        return op1 * op2
    elif op == "/":
        return op1 / op2
    elif op == "+":
        return op1 + op2
    else:
        return op1 - op2


print(postfix_eval('7 8 + 3 2 + /'))
```
